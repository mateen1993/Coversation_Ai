```python
#Import libraries
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
# Load the image for "mask" parameter task
mask_image = np.array(Image.open("/content/WordCloudText/btu logo.png"))
```

```python
# Load the text file
with open("/content/WordCloudText/JaneEyre.txt", "r") as file:
  text = file.read()
```

```python
# Create a WordCloud object

custom_stopwords = set(STOPWORDS)
# Add words to exclude from the word cloud
custom_stopwords.update(["the", ",", "I"])

wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="grey",
    colormap="viridis",
    max_words=50,
    stopwords=custom_stopwords,     # Apply the stopwords list
    mask=mask_image,                # Apply the mask image
    contour_color='blue',           # Outline color around the shape
    contour_width=1,                # Outline thickness
    min_font_size=10, ).generate(text)
```

```python
# Display the word cloud
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

**In a word cloud, certain words appear more prominently because word clouds typically scale the font size of each word based on its frequency in the text. Words that occur more frequently are displayed larger. like "said", "now" , "will", "one" and so on.**
